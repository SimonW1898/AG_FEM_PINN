# Helmholtz Equation Variational Problem

Given the wavenumber $k_0 = 4 \pi$, find $u \in V$ such that

$$
a(u, v) = L(v) \quad \forall v \in V,
$$

where the bilinear form $a(\cdot, \cdot)$ and linear form $L(\cdot)$ are defined by

$$
a(u, v) = \int_\Omega \nabla u \cdot \nabla v \, dx - k_0^2 \int_\Omega u v \, dx,
$$

$$
L(v) = \int_\Omega f v \, dx,
$$

with the source function

$$
f(\mathbf{x}) = A k_0^2 \cos(k_0 x) \cos(k_0 y),
$$

where $A$ is the source amplitude (possibly complex).

---

# Exact solution

The exact solution is

$$
u_\text{exact}(\mathbf{x}) = A \cos(k_0 x) \cos(k_0 y).
$$

---

# Error definitions

Define the error function

$$
e = u_h - u_\text{exact},
$$

where $u_h$ is the finite element approximation.

The relative $H^1$ error is computed as

$$
\text{Relative } H^1\text{-error} = \frac{\| \nabla e \|_{L^2(\Omega)}}{\| \nabla u_\text{exact} \|_{L^2(\Omega)}} = \frac{\left(\int_\Omega |\nabla e|^2 \, dx \right)^{1/2}}{\left(\int_\Omega |\nabla u_\text{exact}|^2 \, dx \right)^{1/2}},
$$

and the relative $L^2$ error is

$$
\text{Relative } L^2\text{-error} = \frac{\| e \|_{L^2(\Omega)}}{\| u_\text{exact} \|_{L^2(\Omega)}} = \frac{\left(\int_\Omega |e|^2 \, dx \right)^{1/2}}{\left(\int_\Omega |u_\text{exact}|^2 \, dx \right)^{1/2}}.
$$


The PDE is:
$$
-\Delta u - k_0^2 u = f \quad \text{in } \Omega,
$$

Modify it to a complex problem by setting $k_0 = 4 \pi i$.

The analytical solution becomes:
$$
u_\text{exact}(\mathbf{x}) = A \cos(k_0 x) \cos(k_0 y).
$$
